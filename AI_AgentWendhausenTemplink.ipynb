```python
# Install required packages
!pip install pandas numpy matplotlib seaborn scikit-learn langchain-openai langchain flask pyngrok --quiet

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from langchain_openai import OpenAI
from langchain.agents import initialize_agent, Tool
import warnings
import os
from google.colab import files, userdata
from flask import Flask, request, render_template_string
from pyngrok import ngrok

warnings.filterwarnings('ignore')

# Securely set OpenAI API key in Colab
try:
    os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
    print("OpenAI API key successfully retrieved from Colab Secrets.")
except Exception as e:
    print("Error: Failed to retrieve OPENAI_API_KEY from Colab Secrets.")
    print("Please set up your API key with the following steps:")
    print("1. In the Colab left sidebar, click the key icon (Secrets).")
    print("2. Click '+ New Secret'.")
    print("3. Enter 'OPENAI_API_KEY' (exactly as shown, case-sensitive) as the Name.")
    print("4. Paste your OpenAI API key (starting with 'sk-') as the Value. Get it from https://platform.openai.com/.")
    print("5. Save the secret and rerun the script.")
    print(f"Error details: {str(e)}")
    raise

# Set ngrok authtoken (already configured, no change needed)
# If not set previously, replace YOUR_AUTHTOKEN_HERE with your actual authtoken
!ngrok authtoken 2zru7MztyqdoKV4i2dbO60iTWiu_6rtFCsbjn4PEBDyKq3o1j  # Ensure this is set as per previous instructions

# Upload dataset to Colab
print("Please upload the 'WendhausenHarvest.csv' file:")
uploaded = files.upload()

# Dynamically determine the uploaded file name
import os
uploaded_files = list(uploaded.keys())
if not uploaded_files:
    raise FileNotFoundError("No file was uploaded.")
latest_file = uploaded_files[0]  # Use the first uploaded file (should be WendhausenHarvest.csv with suffix)

# Load the dataset
try:
    df = pd.read_csv(latest_file)
    print(f"Successfully loaded: {latest_file}")
except FileNotFoundError:
    print(f"Error: Could not load {latest_file}. Ensure the file was uploaded correctly.")
    raise
except Exception as e:
    print(f"Error loading {latest_file}: {str(e)}")
    raise

# Inspect column names
print("Available columns:", df.columns.tolist())

# Map expected columns to actual columns
column_mapping = {
    'dry matter harvest (t/ha)': 'Dyieldtha',  # Mapped to dry yield
    'daily_datetime': 'daily_datetime'        # Direct match
}
for expected, actual in column_mapping.items():
    if actual not in df.columns:
        print(f"Error: Mapped column '{actual}' for '{expected}' not found in dataset.")
        raise KeyError(f"Missing mapped column: {actual}")

# Assign mapped columns
df['dry_matter_harvest'] = df[column_mapping['dry matter harvest (t/ha)']]
df['daily_datetime'] = pd.to_datetime(df[column_mapping['daily_datetime']], errors='coerce')

# Data cleaning and preprocessing
columns_to_drop = ['SystemType', 'TreeType', 'Distance', 'orientation', 'sensor_id', 'WindDirec', 'Gustspeed', 'AnemTemp', 'degrees x-axis level', 'degrees y-axis level', 'VapPress', 'AtmPress', 'VPD', 'RefPressu', 'SMC3', 'STC3', 'SMC4', 'STC4', '1000seedwt', 'hectoliterW', 'CrudeProten', 'CrudeStarch']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Define features and target (adjusted for available columns)
features = [
    'AirTemp',              # degree_c air temperature
    'SRAD',                 # w/m2 solar radiation
    'SMC1',                 # m3/m3 15 cm water content 1
    'STC1',                 # degree_c 15 cm soil temperature 1
    'SMC2',                 # m3/m3 30 cm water content 2
    'STC2',                 # degree_c 30 cm soil temperature 2
    'Windspeed',            # m/s wind speed
]
target = 'dry_matter_harvest'

missing_cols = [col for col in features + [target] if col not in df.columns]
if missing_cols:
    print(f"Error: The following columns are missing from the dataset: {missing_cols}")
    raise ValueError("Missing required columns for modeling.")

model_df = df[features + [target]].dropna()

if model_df.shape[0] < 10:
    print(f"Error: Only {model_df.shape[0]} rows remain after dropping NaNs. Need more data for modeling.")
    raise ValueError("Insufficient data for modeling.")

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(model_df[features])
y = model_df[target]

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

best_rf = grid_search.best_estimator_
print("\nBest Hyperparameters:", grid_search.best_params_)

cv_scores = cross_val_score(best_rf, X_scaled, y, cv=3, scoring='r2')
print(f"Cross-Validation R² Scores: {cv_scores}")
print(f"Average CV R² Score: {cv_scores.mean():.2f} (±{cv_scores.std():.2f})")

y_pred = best_rf.predict(X_test)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
print(f"\nTest Set Evaluation:")
print(f"R² Score: {r2:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")

# Feature importance plot
importances = best_rf.feature_importances_
indices = np.argsort(importances)[::-1]
feature_names = [features[i] for i in indices]

plt.figure(figsize=(10, 6))
sns.barplot(x=importances[indices], y=feature_names)
plt.title("Feature Importances for Yield Prediction")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()

# Predicted vs. Actual scatter plot
plt.figure(figsize=(8, 8))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("Actual Dry Matter Harvest (t/ha)")
plt.ylabel("Predicted Dry Matter Harvest (t/ha)")
plt.title("Predicted vs. Actual Yield")
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(model_df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix of Features and Target")
plt.tight_layout()
plt.show()

# Initialize LLM
try:
    llm = OpenAI(temperature=0)
    print("OpenAI LLM initialized successfully.")
except Exception as e:
    print(f"Error initializing OpenAI LLM: {str(e)}")
    print("Please verify your OpenAI API key and ensure you have sufficient credits.")
    raise

# Define tools for the agent
def get_yield_stats(query):
    """Get detailed statistics about crop yields"""
    if target not in df.columns:
        return "Error: Yield data not found in the dataset."
    stats = df[target].describe()
    result = f"Crop yield statistics (t/ha):\n"
    result += f"Count: {stats['count']:.0f}\n"
    result += f"Mean: {stats['mean']:.2f}\n"
    result += f"Std: {stats['std']:.2f}\n"
    result += f"Min: {stats['min']:.2f}\n"
    result += f"25%: {stats['25%']:.2f}\n"
    result += f"50% (Median): {stats['50%']:.2f}\n"
    result += f"75%: {stats['75%']:.2f}\n"
    result += f"Max: {stats['max']:.2f}\n"

    sample_df = df[target].sample(frac=0.1, random_state=42)
    plt.figure(figsize=(8, 6))
    sns.histplot(sample_df, bins=30, kde=True)
    plt.title("Distribution of Dry Matter Harvest (t/ha)")
    plt.xlabel("Dry Matter Harvest (t/ha)")
    plt.ylabel("Frequency")
    plt.show()

    return result

def get_weather_impact(query):
    """Analyze impact of weather on yields"""
    weather_cols = ['AirTemp', 'SRAD', 'Windspeed', target]
    if not all(col in df.columns for col in weather_cols):
        return f"Error: Some required columns ({weather_cols}) are missing."
    corr = df[weather_cols].corr()
    result = f"Weather-yield correlations:\n{corr.to_string()}\n\n"
    target_corr = corr[target].drop(target)
    for feature, value in target_corr.items():
        result += f"- {feature}: {value:.2f} ("
        if abs(value) > 0.5:
            result += "strong "
        elif abs(value) > 0.3:
            result += "moderate "
        else:
            result += "weak "
        result += "correlation)\n"

    sample_df = df[weather_cols].sample(frac=0.1, random_state=42)
    plt.figure(figsize=(8, 8))
    sns.heatmap(sample_df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title("Correlation Matrix of Weather Features and Yield")
    plt.show()

    return result

def predict_yield(params):
    """Predict yield based on environmental factors"""
    try:
        features_order = [
            'AirTemp',              # degree_c air temperature
            'SRAD',                 # w/m2 solar radiation
            'SMC1',                 # m3/m3 15 cm water content 1
            'STC1',                 # degree_c 15 cm soil temperature 1
            'SMC2',                 # m3/m3 30 cm water content 2
            'STC2',                 # degree_c 30 cm soil temperature 2
            'Windspeed',            # m/s wind speed
        ]
        values = params.replace('(', '').replace(')', '').split(',')
        values = [float(v.strip().split()[0]) for v in values if v.strip()]
        if len(values) != len(features_order):
            return f"Please provide exactly {len(features_order)} numerical parameters in order: {', '.join(features_order)}"
        input_df = pd.DataFrame([values], columns=features_order)
        input_scaled = scaler.transform(input_df)
        prediction = best_rf.predict(input_scaled)[0]
        return f"Predicted yield: {prediction:.2f} t/ha"
    except Exception as e:
        return f"Error: Could not process your request. Please provide numerical parameters for all features in order: {', '.join(features_order)}. Details: {str(e)}"

# Define tools
tools = [
    Tool(
        name="YieldStats",
        func=get_yield_stats,
        description="Provides detailed statistics and a histogram for crop yields (dry matter harvest in t/ha)."
    ),
    Tool(
        name="WeatherImpact",
        func=get_weather_impact,
        description="Analyzes how weather factors (temperature, solar radiation, wind speed) affect crop yields, including correlations and a heatmap."
    ),
    Tool(
        name="YieldPredictor",
        func=predict_yield,
        description=f"Predicts crop yield based on environmental factors. Provide {len(features)} numerical values in order: {', '.join(features)}."
    )
]

# Initialize agent
try:
    agent = initialize_agent(tools, llm, agent="zero-shot-react-description", verbose=True)
    print("LangChain agent initialized successfully.")
except Exception as e:
    print(f"Error initializing LangChain agent: {str(e)}")
    print("Please verify your OpenAI API key and ensure you have sufficient credits.")
    raise

# Example queries
print("\nRunning example queries:")
print("\n1. Yield Statistics:")
try:
    print(agent.run("What are the statistics for crop yields?"))
except Exception as e:
    print(f"Error running YieldStats query: {str(e)}")

print("\n2. Weather Impact:")
try:
    print(agent.run("How does temperature affect yields?"))
except Exception as e:
    print(f"Error running WeatherImpact query: {str(e)}")

print("\n3. Yield Prediction:")
example_params = "15, 200, 0.3, 10, 0.25, 12, 2"  # Adjusted for 7 features
try:
    print(agent.run(f"Predict yield for {example_params}"))
except Exception as e:
    print(f"Error running YieldPredictor query: {str(e)}")

# Web deployment setup with Flask and ngrok
app = Flask(__name__)

@app.route('/')
def home():
    html = '''
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Crop Yield Dashboard</title>
        <style>
            body { font-family: Arial, sans-serif; margin: 20px; }
            .input-group { margin-bottom: 15px; }
            label { display: inline-block; width: 200px; }
            button { margin-right: 10px; }
            #output { margin-top: 20px; }
        </style>
    </head>
    <body>
        <h1>Crop Yield Dashboard</h1>
        <p>Enter weather data and click a button to see results!</p>
        <div class="input-group">
            <label>Air Temperature (°C): <input type="number" id="temp" value="15" step="0.1" title="Enter temperature in °C"></label><br>
            <label>Solar Radiation (W/m²): <input type="number" id="solar" value="200" step="0.1" title="Enter solar radiation in W/m²"></label><br>
            <label>Water Content 15cm (m³/m³): <input type="number" id="water15" value="0.3" step="0.01" title="Enter water content at 15cm depth"></label><br>
            <label>Soil Temp 15cm (°C): <input type="number" id="soil15" value="10" step="0.1" title="Enter soil temperature at 15cm"></label><br>
            <label>Water Content 30cm (m³/m³): <input type="number" id="water30" value="0.25" step="0.01" title="Enter water content at 30cm depth"></label><br>
            <label>Soil Temp 30cm (°C): <input type="number" id="soil30" value="12" step="0.1" title="Enter soil temperature at 30cm"></label><br>
            <label>Wind Speed (m/s): <input type="number" id="wind" value="2" step="0.1" title="Enter wind speed in m/s"></label><br>
        </div>
        <button onclick="getYieldStats()">Get Yield Stats</button>
        <button onclick="getWeatherImpact()">Show Weather Impact</button>
        <button onclick="predictYield()">Predict Yield</button>
        <div id="output"></div>

        <script>
            function getYieldStats() {
                fetch('/yield_stats').then(response => response.text()).then(data => {
                    document.getElementById('output').innerText = data;
                });
            }

            function getWeatherImpact() {
                fetch('/weather_impact').then(response => response.text()).then(data => {
                    document.getElementById('output').innerText = data;
                });
            }

            function predictYield() {
                const params = [
                    document.getElementById('temp').value,
                    document.getElementById('solar').value,
                    document.getElementById('water15').value,
                    document.getElementById('soil15').value,
                    document.getElementById('water30').value,
                    document.getElementById('soil30').value,
                    document.getElementById('wind').value
                ].join(',');
                fetch(`/predict?params=${params}`).then(response => response.text()).then(data => {
                    document.getElementById('output').innerText = data;
                });
            }
        </script>
    </body>
    </html>
    '''
    return render_template_string(html)

@app.route('/yield_stats')
def yield_stats():
    return get_yield_stats("")

@app.route('/weather_impact')
def weather_impact():
    return get_weather_impact("")

@app.route('/predict')
def predict():
    params = request.args.get('params')
    return predict_yield(params)

# Start Flask app with ngrok
public_url = ngrok.connect(5000).public_url
print(f"Public URL: {public_url}")
app.run(host='0.0.0.0', port=5000)
```
